Start with a Dataframe, then group by something(step 2) step 3 is applying a function ie. max/mean, then the result is the new dataframe.

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
#using seaborn to import data set from the titanic
```

```python
titanic.head(50)
#loading dataset
```

```python
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean')
#taking the mean of survivors by sex and class
```

```python
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
#removes verical alignment and makes it horizontal and cleaner
```

```python
titanic.pivot_table('survived', index='sex', columns = 'class')
#creating pivot table to display the information (same as groupby but less work)
```

```python
titanic.pivot_table('survived', 'sex', 'class')
#shows that entering the index and columns are optional, good to use index and columns if you are newer
```

```python
titanic.pivot_table('age', ['sex', 'survived'], 'class')
# making the PT multidimensional by passing a list to one of the arguments with multiple variables
```

```python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', 'age'], 'class')
```

```python
titanic.pivot_table(index='sex', columns = 'class',
                    aggfunc={'survived': sum, 'fare': 'mean'})
#using aggfunc to specify what fucntions I want ran and on what variable
```
